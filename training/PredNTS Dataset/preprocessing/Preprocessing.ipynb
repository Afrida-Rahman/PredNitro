```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('http://kurata14.bio.kyutech.ac.jp/PredNTS/download_file/Training-datasets-PredNTS.txt', sep='\t', names=['Sequence','protein_id','position','residue','label','feature'])
```

```python
dataset.to_csv('training-dataset-PredNTS.csv',index=None)
```

```python
df=dataset['Sequence'].str.replace('-', 'X')
df.head(23)
```

```python
dataset[['Sequence']] = df
dataset.head(23)
```

```python
positive_data=dataset.loc[dataset['label']==1]
df_p=positive_data['Sequence']
df_p.to_csv('pos.txt',index=None,header=None)
```

```python
negative_data=dataset.loc[dataset['label']==-1]
df_n=negative_data['Sequence']
df_n.to_csv('neg.txt',index=None,header=None)
```
